```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
```

```python
#Simulate data
n = 1000
x = np.random.normal(3, 5, size=n)
y = -15*x + 20 + np.random.normal(0, 3, size=n)
```

```python
#Analytcaly solve linear regressioon
X = np.sum(x)
Y = np.sum(y)
b1 = (n*np.dot(x,y) - X*Y)/(n*np.dot(x,x)-X**2)
b2 = (Y-b1*X)/n
b1,b2 = round(b1, 2), round(b2, 2)
print("y = ",b1," x +",b2)
```

```python
def MSE_loss(b0, b1, x, y):
    pred = b0 + b1 * x
    loss = np.sum((pred - y) ** 2) / len(y)
    return loss

# Create a grid of b0 and b1 values around the actual b0=20 and b1=-15.
b0_val = np.linspace(0,40,100)
b1_val = np.linspace(-30,0,100)

# Calculate MSE loss for each point of this grid
# Log-scale is used only for better visualization
Z = np.zeros((len(b0_val),len(b1_val)))
for i in range(len(b0_val)):
    for j in range(len(b1_val)):
        Z[i,j] = np.log(MSE_loss(b0_val[i], b1_val[j],x,y))

# 3D Plot of MSE loss in log-scale
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111,projection='3d')
B0, B1 = np.meshgrid(b0_val, b1_val)
ax.plot_surface(B0, B1, Z, cmap='viridis')


# Adding labels
ax.set_xlabel('b0')
ax.set_ylabel('b1')
ax.set_zlabel('Loss', labelpad=15)
ax.zaxis.label.set_rotation(90)  

plt.show()
```
